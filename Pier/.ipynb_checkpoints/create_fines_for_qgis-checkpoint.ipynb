# Create a csv for Qgis dividing the fines dataset in a grid of 20x20 bins based on the gps data and summing the fines for each bin 

```python
import warnings, numpy as np, pandas as pd, pymysql, matplotlib, matplotlib.pyplot as plt, seaborn as sb, geopandas as gp, geopy
from datetime import datetime, timedelta
```

```python
multe = pd.read_csv("data/fines.csv", parse_dates=["date"])
multe = multe.dropna()
multe = multe.loc[:, ["date", "lat", "long", "num"]].copy()
multe.head()
```

```python
multe['long'] = pd.to_numeric(pd.cut(multe['long'], bins=20).apply(lambda x: (x.left+x.right)/2))
multe['lat'] = pd.to_numeric(pd.cut(multe['lat'], bins=20).apply(lambda x: (x.left+x.right)/2))
```

```python
multe.head()
```

```python
print(multe.size)
multe = multe.groupby(["date","long","lat"]).agg({"num":'sum'}).reset_index()
multe = multe.set_index(["date","long","lat"], drop=True).reset_index()

print(multe.size)
multe
```

```python
multe.to_csv("data/fines_for_qgis.csv")
```
