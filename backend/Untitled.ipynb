```python
import pymysql
```

```python
connection = pymysql.connect("194.116.76.192", "bd7", "bd7", "final_project")
cursor = connection.cursor(pymysql.cursors.DictCursor)
```

```python
query = '''
    SELECT * 
    FROM trips_parks_auto_concerto
    ORDER BY device_id,  event_date
    '''

print('eseguo query')

cursor.execute(query)
rv = cursor.fetchall()

print('query eseguita')

    
```

```python
# Calcolo set auto univoche e timestamp di inizio
dist = set()
min_tts = 1000000000
for row in rv:
    dist.add(row['device_id'])
    if int(row['tts']) < min_tts:
        min_tts = int(row['tts'])

print('set creato')
print(len(dist))

    
```

```python
import random
# Creo il JSON nel formato richiesto da deck.gl
results = []
for car in dist:
    old_lat = 0
    old_long = 0
    print('Elaboro auto ' + str(len(results)) + ' su ' + str(len(dist)))

    color = [159, 229, 90]

    dati_riga = {
        'car_id': car,
        'color': color,
        'segments': []
        }
    for row in rv:
        if row['device_id'] == car:
            
            check = False

            lat = row['latitude']
            long = row['longitude']
            
            if old_lat == 0:
                old_lat = lat
                old_long = long
                check = True
                
            if not check:
                if (lat >= old_lat - (old_lat)*0.002 and lat <= old_lat + (old_lat)*0.002) and\
                (long >= old_long - (old_long)*0.002 and long <= old_long + (old_long)*0.002):
                    check = True
                else:
                    print('errore')
            if row['status'] == 'parked':
                lat = lat + random.uniform(-0.0001, 0.0001)
                long = long + random.uniform(-0.0001, 0.0001)

            punti = [long, lat, int(row['tts']) - min_tts, row['status']]
            
            if check:
                dati_riga['segments'].append(punti)

    results.append(dati_riga)

print('lista creata')
```

```python
import json
with open('concerto.json', 'w') as outfile:
    json.dump(results, outfile)
```
